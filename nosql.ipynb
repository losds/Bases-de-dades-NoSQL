```python
#Muestro solo aprobados con la nota y curso
myquery = { "mark": { "$gt": "5" } }
mydoc = mycol.find(myquery,{ "_id": 0, "mark": 1, "course": 1 })
for x in mydoc:
    print(x) 
```

```python
#orden descendente
mydoc = mycol.find({},{ "_id": 0, "mark": 1, "course": 1 }).sort("mark", -1)

for x in mydoc:
  print(x) 
```
